# CityBikes

Send a request to CityBikes for the city of your choice.

```python
import pandas as pd
import requests

#Send Get Request
url= "https://api.citybik.es/v2/networks/bixi-toronto"

#Parsing JSON Response
response = requests.get(url)

data = response.json()
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes).

```python
toronto_stations = []

for station in data['network']['stations']:
  station_info = {
        "Station": station["name"],
        "Latitude": station["latitude"],
        "Longitude": station["longitude"],
        "Available Bikes": station["free_bikes"]
    }
  toronto_stations.append(station_info)

print(toronto_stations)
```

Put your parsed results into a DataFrame.

```python
toronto_city_bikes_df = pd.DataFrame(toronto_stations)
toronto_city_bikes_df
```

```python
#saving the data into a CSV file
toronto_city_bikes_df.to_csv("../data/toronto_city_bikes_stations.csv", index=False)
```
